```python
# Identifying the top 20 most occured qualification and creating dummy variables for those 20 qualifications:

most_occ = sorted(data['Qualification'].items(), key = lambda x:x[1],reverse=True)[:20]

final_qual = []

for tup in most_occ:
    final_qual.append(tup[0])
    
for title in final_qual:
    data[title]=0
    
for x,y in zip(data["Qualification"].values, np.array([idx for idx in range(len(data))])):
    for q in x:
        q = q.strip()
        if q in final_qual:
            data[q][y] = 1
```

```python
# Identifying the top 20 most occured qualification and creating dummy variables for those 20 qualifications:

most_occ = sorted(data_test['Qualification'].items(), key = lambda x:x[1],reverse=True)[:20]

final_qual = []

for tup in most_occ:
    final_qual.append(tup[0])
    
for title in final_qual:
    data_test[title]=0
    
for x,y in zip(data_test["Qualification"].values, np.array([idx for idx in range(len(data_test))])):
    for q in x:
        q = q.strip()
        if q in final_qual:
            data_test[q][y] = 1
```
