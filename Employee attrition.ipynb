ONE HOT ENCODING

```python
data=pd.get_dummies(data=data_set, drop_first=True)
data.shape
```

```python
len(data.select_dtypes(include='object').columns)
```

Earlier the total columns were 31 and after encoding columns are 45. We can see that no categorical columns are left in our data set.

```python
pd.set_option("display.max_columns",None)
data.head()
```

Now, here we can see that all the columns are numerical.

```python
# renaming target variable
data.rename(columns={'Attrition_Yes': 'Attrition'}, inplace=True)
data.head()
```

SPLITTING THE DATASET

```python
# matrix of features

x = data.drop(columns = 'Attrition')
```

```python
# target variable

y = data['Attrition']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

FEATURE SCLAING

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

BUILDING THE (CLASSIFICATION) MODEL

```python
# 1: Logistic Regression
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()

lg.fit(x_train, y_train)
```

```python
y_pred = lg.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred) 
print(accuracy*100)
```

```python
conm= confusion_matrix(y_test, y_pred)
print(conm)
```

```python
# 2: random forest

from sklearn.ensemble import RandomForestClassifier
```

```python
cl = RandomForestClassifier(random_state=0)
cl.fit(x_train, y_train)
```

```python
y_pred= cl.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred)
print(acc*100)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# 3. Support Vector Machine

from sklearn.svm import SVC
```

```python
svc = SVC(random_state=0)
svc.fit(x_train, y_train)
```

```python
y_predc = svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accur = accuracy_score(y_test, y_predc)
print(accur*100)
```

```python
cm1 = confusion_matrix(y_test, y_predc)
print(cm1)
```
